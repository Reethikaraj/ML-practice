### Q1.
* Suppose we have a discrete random variable X that can take on the values 1, 2, 3, and 4 with probabilities 0.1, 0.4, 0.3, and 0.2 respectively.
* Write the probability mass function (PMF) for X.
* Calculate the cumulative distribution function (CDF) for X.


```python
probabilities = {1: 0.1, 2: 0.4, 3: 0.3, 4: 0.2}
pmf = probabilities
cdf = {}
cumulative_prob = 0
for x in sorted(probabilities.keys()):
    cumulative_prob += probabilities[x]
    cdf[x] = cumulative_prob

print("PMF for X:", pmf)
print("CDF for X:", cdf)
```
