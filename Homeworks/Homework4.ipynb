### Q1. 


Suppose we want to test whether the mean weight of apples in a grocery store is 150 grams. We randomly sample 20 apples from the store and measure their weights, getting the following data:

Apple_weights = [145, 155, 160, 146, 142, 152, 150, 147, 148, 149, 148, 152, 153, 155, 154, 148, 151, 147, 153, 146]

* What test should we use and why?
* State the null and alternative hypotheses.
* Choose a significance level (α) (the probability of rejecting the null hypothesis when it is actually true).
* Determine the degrees of freedom (df) of the sample.
* Determine the critical value of t based on the significance level and degrees of freedom. For a two-tailed test with α = 0.05 and df = 19, the critical values are -2.093 and 2.093.
* Compare and interpret the results of the test to the critical value



#### 1.1 What test should we use and why?

Observations of the provided data:

    1. Input variables are discrete. 
    2. Sample size is 20. Which is less than 30.
    3. Population mean is unknown.
So, with all the above observations, we need to perform two tailed t-test.

t = (x̄ - μ) / (s / sqrt(n))

where 
* x̄ is the sample mean
* μ is the hypothesized population mean 
* s is the sample standard deviation
* n is the sample size.

#### 1.2 State the null and alternative hypotheses.
    1. Null Hypothesis: Mean weight of apples in a grocery store is 150 grams
    2. Alternate Hypothesis: Mean weight of apples in a grocery store is not 150 grams. (It could be higher or lower).


#### 1.3 Choose a significance level (α) (the probability of rejecting the null hypothesis when it is actually true).

```python
# Significance Level alpha 
alpha = 0.05
```

#### 1.4 Determine the degrees of freedom (df) of the sample.

```python
apple_sample_data = np.array([145, 155, 160, 146, 142, 152, 150, 147, 148, 149, 148, 152, 153, 155, 154, 148, 151, 147, 153, 146])
n = len(apple_sample_data)
degree_of_freedom = n-1
degree_of_freedom
```

#### 1.5 Determine the critical value of t based on the significance level and degrees of freedom

```python
import random
import numpy as np
import scipy.stats as stats
```

```python
# Hypothesized population mean
pop_mu = 150

sample_mean = apple_sample_data.mean().round(2)
sample_std = apple_sample_data.std().round(2)

# Calculating t_stat
t_stat = (sample_mean - pop_mu) / (sample_std / np.sqrt(n))
print(f"t-statistic: {t_stat:.4f}")

t_critical_high = stats.t.ppf(alpha/2, degree_of_freedom)  
t_critical_low = -stats.t.ppf(alpha/2, degree_of_freedom)
print(f"t-critical: {t_critical_low:.4f}, {t_critical_high:.4f}")

p_value = stats.t.sf(np.abs(t_stat), degree_of_freedom) * 2  
print(f"p-value: {p_value:.4f}")

if (t_stat < -2.093 or t_stat > 2.093):
    print('Reject the null hypothesis that says that the population mean = %.3f'%pop_mu)
else:
    print('Fail to reject or accept the null hypothesis, the population mean = %.3f'%pop_mu)
```

#### 1.6 Compare and interpret the results of the test to the critical value

The result is true since when p-value is also greater than alpha, we fail to reject the null hypothesis.
