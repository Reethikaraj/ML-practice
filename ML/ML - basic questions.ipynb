# Introduction

Q. What is Machine learning?

     Machine learning (ML) is a subset of AI which provides machines the ability to learn automatically and improve from the experience without being explicitly programmed. It uses statistical methods to enable machine to improve with experience.
     
     Machine learning algorithms build a model based on sample data, known as training data, in order to make predictions or decisions without being explicitly programmed to do so.

Q. What are steps of ML?

      1. Collection of data
      2. Data preparation or cleaning - Handling missing data and outliers
      3. Choosing ML model
      4. Training the model
      5. Evaluating the model
      6. Parametr tuning - going back and handling missing data in a different way
      7. Make predictions

Q. What are types of Machine learning?

    1. Supervised - Machine learns under guidance. Feeding them labelled data. eg. Child
    
    2. Unsupervised - Data is not labelled. Machine has to figure out hidden patterns and make conclusions. eg. adult
    
    3. Reinforcement - Establish and encourage a pattern or behaviour. Hit and trial. Learn from experiences by interacting with environment. eg. We get lost in an island.
