## Simple Linear Regression

Q. Simple linear regression

    y = mx+c (m = slope) : 1 independent variable, 1 dependent variable, 1 constant. 
    
    ML algorithm tries to draw the best possible graph with the given data. It draws a line such that difference between the observed values and predicted values is the least.  
    
Q. Real world applications:

    There are no real world applications for simple linear regression as it has very less number of independent variables and any real world application will depend upon more than 1 variable. There comes the necessity of multiple linear regression. 

# Predicting salary based on years of experience

## Importing libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the data

```python
data = pd.read_csv("salary_SLR.csv")
data
```

```python
# Independent variable is years of experience
# Storing the independent variable into X (All columns except the last)

X = data.iloc[:,:-1].values
X
```

```python
# Dependent variable is salary. Storing that into y

y = data.iloc[:,-1].values
y
```

## Splitting training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

## Training the model using Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Prediction 

```python
y_pred = regressor.predict(X_test)
```

## Visualization of taining data

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title("Experience vs CTC")
plt.xlabel("Years of Experience")
plt.ylabel("CTC")
plt.show()
```

## Visualization of testing data

```python
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_test, regressor.predict(X_test), color="blue")
plt.title("Experience vs CTC")
plt.xlabel("Years of Experience")
plt.ylabel("CTC")
plt.show()
```

## Implementation of multiple linear regression

### Predicting antibodies based on given factors and conditions

Antibodies - Dependent variable

Age, Chronic diseases, body temperature, Blood O2 level - Independent variables

### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing data

```python
data = pd.read_csv("Covid_data Multiple Linear Regr.csv")
data
```

```python
X = data.iloc[:,:-1].values
X
```

```python
y = data.iloc[:,-1].values
y
```

### Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

### Splitting Training and Testing Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=42)
```

### Training the multiple linear regression model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Predicting the test set results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision = 3)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test),1)), 1))
```

### Visualizing results

```python
plt.plot(y_pred)
plt.plot(y_test)
plt.show()
```
