## Addition rules (Probability, "or")

Q. What is mutual exclusion?

     Two events are mutualt exclusive if they cannot occur simultaneously.
     eg: heads and tails cannit happen simultaneously
     
    
Q. What is addition rule of mutually exclusive events?
   
    P(A or B) = P(A) + P(B)
    
    eg: P(1 or 2 or 6) while rolling a dice
    P(1 or 2 or 6) = P(1) + P(2) + P(6) = 1/6 + 1/6 + 1/6 = 3/6 = 0.5

Q. What is addition rule of non mutually exclusive events?
    
    P(A or B) = P(A) + P(B) - P(A and B)


## Conditional Probability (Baye's theorem, "and")

Q. What is Conditional probability for independent events?  

     P(A and B) = P(A)*P(B)
     
Q. What is Conditional probability for dependent events?  

     P(A and B) = P(A)*P(B/A)
     eg: P of getting Queen and Ace in a deck of cards:
     P(Q and A) = P(Q) * P(A)
                = 4/52 * 4/51 (Total no. of cards become 51 since we have already picked up queen)

Entropy - Measure of randomness in a dataset.

* Entropy is directly proportional to impurity.
* Information gain is inversely proportional to impurity

        entropy = 1, balanced data
        entropy < 1, imbalanced data
        entropy = 0, only one class of data, pure data

        Parent = complete data set
        child = random subset of the data set
Information gain is more about the importance of feature of the data. To find which feature has more impact on the decision process.

```python
import numpy as np
from math import log2

def entropy(probabilities):
    E=0
    for p in probabilities:
        E += p+log2(p)
#         print(f"Entropy: {E}")
    return E

def information_gain(E,weights,entropies):
    return E - np.sum(np.multiply(weights,entropies))


#let's consider a dataset of 30 items with 0 (16 items),1(14 items)
parent = [0]*16 + [1]*14
#Let's divide the dataset into 2 random subsets
child1 = [0]*4 + [1]*13 #total 17
child2 = [0]*12 + [1]*1 #total 13

E_parent = entropy([16/30,14/30])
E_child1 = entropy([4/17,13/17])
E_child2 = entropy([12/13, 1/13])

#weight = sizeOfChild/parentSize
information_gain = information_gain(E_parent,[17/30,13/30],[E_child1,E_child2])

print(f"Parent-Entropy: {E_parent} \nChild1-Entropy: {E_child1} \nChild2-Entropy: {E_child2} \nInformation-gain: {information_gain}")
```

When we are slpitting the data, it's parent to children and the information loss should be minimum. When we go from children to parent, the information gain shoul be maximum.
