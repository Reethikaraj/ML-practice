No. of dimensions of the plot = no. of features of the data

* 1 feature - univariate analysis - distplot
* 2 features - Bi variate analysis - jointplot
* 3 or more features - Pairplot / Scatterplot


From 3d plots, it is difficult to visualize. Seaborn is more helpful for this.


## Distribution plots - Continuous variables

* joinplot - Bivariate analysis - For 2 features
* pairplot - Scatterplot -  more than 2 features
* distplot - According to density or numbers

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips") 
#tips is a dataset that is already present inside seaborn
df.head()
```

Dependent feature - tip

All other are independent features

### Correlation with Heatmap

It uses colored cells, typically in a monochromatic scale, to show a 2D correlation between 2 discrete dimensions or event types. It is a very important feature selection.

Correlation is only for int or float types. Not for categorical variables(since they are object type.)

Correlation varies between -1 and +1 (Pearson correlation)

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

### Joinplot 
Helps in bivariate (Only 2 features)

```python
sns.jointplot(x="tip", y="total_bill", data = df, kind="hex")
```

```python
sns.jointplot(x="tip", y="total_bill", data = df, kind="reg")
```

### Pairplot - Scatterplot

If we have more than 2 features. 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex")
# For categorical variables
```

```python
sns.pairplot(df, hue="smoker")
# For categorical variables
```

### Dist plot

For histogram

```python
sns.distplot(df["tip"])
# kde - Kernel density estimation - true=density (default)
# bins - numbers
```

```python
sns.distplot(df["tip"], kde=False, bins=10)
```

## Categorical plots - Categorical variables
 * countplot - 1 feature
 * barplot - 2 features
 * boxplot - To find percentile and mean - 3 features, x,y,hue
 * violinplot - Combination of all with more than 2 features. 3 features, x,y,hue.
 

### Countplot
 We need to give x or y value and data

```python
sns.countplot(x="smoker", data=df)
```

```python
sns.countplot(x="day", data=df)
```

```python
sns.countplot(y="day", data=df)
```

### Barplot
 We need to give x and y values and data

```python
sns.barplot(x="smoker", y="total_bill", data=df)
```

```python
sns.barplot(x="sex", y="total_bill", data=df)
```

### Box plot

```python
sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill',hue='smoker', data=df)
```

### Violin plot

It shows the distribution of the data based on kernel density and box plot.

```python
sns.violinplot(x='day', y='total_bill',hue='smoker', data=df)
```
