Q. What is matlplotlob?

    Library that is used to plot mathematical graphs. 

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

x1 = [1,2,3,4,5]
y1 = [5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x1,y1)
plt.plot(x,y)
plt.plot(x1,y1)

plt.show()
```

```python
plt.plot(x,y, "r") #Change color to red
plt.plot(x1,y1, "g--.") #cHange line pattern and color
plt.show()
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6])
y = 2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1) #Creates an array in range of 0,10 with steo 0.1
y = 2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
plt.plot(x,y, color = "red", marker = "p", linewidth = 3)  # marker = pentagon

plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
y2 = x**3
plt.plot(x,y, color = "red", label = "x**4")  
plt.plot(x,y2, color = "blue", label = "x**3")

#labels for axis
plt.ylabel("y", fontsize = 15)
plt.xlabel("x")

#title
plt.title("Example")

#to show grid
plt.grid()

#to change x axis division
plt.axis([0,15,0,20000])

#to add some text inside the plot
plt.text(10,12500,"Practice",fontsize=12)

plt.legend() #To show the labels
```

## Pie chart

```python
profit = [10,20,30,25]
my_labels = ["Amazon", "Flipkart", "Snapdeal", "Jiomart"]
my_colors = ["red","green","pink","blue"]

#autopct to show percentage and by how many decimal points
#explode = coming out
# pie chart starts at 3o'clock angle. So, amazon is the 1st part. Order is Amazon, Flipkart, Snapdeal, Jiomart
# we can change the direction and angle 
#counterclockwise = False
# startangle = 90

plt.pie(profit, colors = my_colors, labels = my_labels, autopct = "%.2f%%", explode = [0.05,0.1,0,0] )
plt.show()
```
