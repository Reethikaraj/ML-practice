#### Technical terms in database

* NoSQL = Non relational database, to store large data, No security.
* SQL = Relational database = *** Focus of the course, ACID principles, query large database would slow down the system.
* In memory = Small but quicker to relieve
* disk-based = Large


Transaction = Several operation in one go

Entity - Place where we keep several data records - user entity, product entity, order entity

Hierarchy of the database: (low to high)
1. Single record 
2. Entity - Collection (NoSQL), Table(SQL) - 1 primary key per entity 
3. Schema
4. Database
5. Collection of Database
