```python
import requests as req
from bs4 import BeautifulSoup

#send request, get response
response = req.get("https://www.integrify.io/")

#get HTML document response
html_text = response.text

print('Content-type -- ',response.headers.get('Content-Type'))
# print(html_text)

doc = BeautifulSoup(html_text,"html.parser")
paragraphs = doc.find_all('p')

for p in paragraphs:
    print('-----------------------------')
    print(p.get_text())
```

```python
layout = doc.find_all("div", class_="layout-container")
# print(layout)
for item in layout[0].find_all("div", class_='testing'):
    print(item.find('div', class_='paragraph-bigger').get_text())
    print(item.find('div', class_='integrify-in-numbers-paragraph').get_text())
    
```
