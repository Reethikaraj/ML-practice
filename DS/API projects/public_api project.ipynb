Working with the above api which is the list of all open apis

```python
# https://api.publicapis.org/entries

import requests

public_response = requests.get('https://api.publicapis.org/entries')
public_response
```

```python
type(public_response)
```

```python
public_response.status_code
```

```python
public_response.text
```

```python
header = public_response.headers
header
```

```python
header['Date']
```

```python
header['Transfer-Encoding']
```

```python
import json
api_data = public_response.json()
api_data
```

```python
entry_data = api_data['entries']
entry_data
```

```python
for item in entry_data:
    print(item['API'] + ": " + item['Category'])
```

```python
import pandas as pd
df = pd.DataFrame()

for item in entry_data:
    df = df.append({'API Name': item['API'],
              'Category name': item['Category']}, ignore_index=True)
df
```
