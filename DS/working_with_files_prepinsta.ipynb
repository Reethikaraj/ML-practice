## Text files

#### Current working directory:

```python
import os
os.getcwd() #To get the current working directory
```

#### Opening a file:

```python
# open() is an inbuilt function in python to open a file
# This will open the file in read mode. 
sample_text = open('sample.txt')
```

#### Read a file:

```python
text = sample_text.read() 
text
```

```python
#readlines() can also be used to raed multiple lines
sample_text = open('sample.txt')
lines = sample_text.readlines()
lines
```

#### Close a file:

```python
sample_text.close()
```

### For closing a file automatically: "with"

```python
with open('sample.txt') as sample_text:
    print(sample_text.read())
```

### Write into a file:

```python
# write mode will overwrite old data with the new data if there is an existing file by that name.
# If there is no prior existing file by that name, it creates a new file.

fw = open('write_sample.txt', 'w') # write mode
fw.write('This is to test the write mode.')
fw.close()
```

```python
fa = open('write_sample.txt', 'a') # append mode
fa.write('This is to test the append mode.')
fa.close()
```

```python
f = open('write_sample.txt') # default read mode
text = f.read()
f.close()
text
```

### Delete a file:

```python
# we need to import os to use remove method
# import os
os.remove('write_sample.txt')
```

```python
# If the file doesn't exist, the remove method will raise an error. So, better to use it this way:

if os.path.exists('write_sample.txt'):
    os.remove('write_sample.txt')
else:
    print("The file doesn't exist")
```

```python
# with utf-8

data = open('example_movie.csv', encoding = 'utf-8')
csv_data = csv.reader(data)
data_lines = list(csv_data)
data_lines
```

```python
data_lines[:3] # 0 to 2 rows
```

```python
for line in data_lines[:5]: #from 0 to 4 lines
    print(line)
```

```python
len(data_lines)
```

```python
all_titles = []
for line in data_lines[1:15]:
    all_titles.append(line[1])
all_titles
```

```python
director_and_actor = []
for line in data_lines[1:15]:
    director_and_actor.append(line[4] + ''+ line[5])
director_and_actor
```

## Writing to csv file:

```python
output_file = open('new_csv_file.csv', 'w', newline = '')
write_data = csv.writer(output_file, delimiter = ',') # delimeter is to mention how we want to separate the values
write_data.writerow(['a','b','c'])
write_data.writerows([[1,2,3],[4,5,6]])
output_file.close()
```

```python
f = open('new_csv_file.csv', 'a', newline = '')
write_data = csv.writer(f)
write_data.writerow(['new', 'new', 'new'])
f.close()

f1 = open('new_csv_file.csv', 'r')
text = f1.read()
print(text)
f1.close()
```

```python
import os
if os.path.exists('new_csv_file.csv'):
    os.remove('new_csv_file.csv')
else:
    print("The file doesn't exist")
```
