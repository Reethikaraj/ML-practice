### Write into a file:

```python
# write mode will overwrite old data with the new data if there is an existing file by that name.
# If there is no prior existing file by that name, it creates a new file.

fw = open('write_sample.txt', 'w') # write mode
fw.write('This is to test the write mode.')
fw.close()
```

```python
fa = open('write_sample.txt', 'a') # append mode
fa.write('This is to test the append mode.')
fa.close()
```

```python
f = open('write_sample.txt') # default read mode
text = f.read()
f.close()
text
```

## Writing to csv file:

```python
output_file = open('new_csv_file.csv', 'w', newline = '')
write_data = csv.writer(output_file, delimiter = ',') # delimeter is to mention how we want to separate the values
write_data.writerow(['a','b','c'])
write_data.writerows([[1,2,3],[4,5,6]])
output_file.close()
```

```python
f = open('new_csv_file.csv', 'a', newline = '')
write_data = csv.writer(f)
write_data.writerow(['new', 'new', 'new'])
f.close()

f1 = open('new_csv_file.csv', 'r')
text = f1.read()
print(text)
f1.close()
```

```python
import os
if os.path.exists('new_csv_file.csv'):
    os.remove('new_csv_file.csv')
else:
    print("The file doesn't exist")
```

# Working with json:
json - Javascript object notation. 

It is a stringified javascript object or python dictionary.

Any json format is a string format.

```python
# dictionary

person_dct= {
    "name":"Rushikesh",
    "country":"India",
    "city":"Pune",
    "skills":["Python", "Machine Learning"]
}


# JSON: A string form a dictionary
person_json = "{'name': 'Rushikesh', 'country': 'India', 'city': 'Pune', 'skills': ['Python', 'Machine Learning']}"


# we use three quotes and make it multiple line to make it more readable
person_json = '''{
    "name":"Rushikesh",
    "country":"India",
    "city":"Pune",
    "skills":["Python","Machine Learning"]
}'''
```

```python
person_dct
```

```python
person_json
```

### Changing json to dictionary:

```python
import json

# JSON
person_json = '''{
    "name":"Rushikesh",
    "country":"India",
    "city":"Pune",
    "skills":["Python","Machine Learning"]
}'''

# let's change JSON to dictionary - json.loads() method
person_dct = json.loads(person_json)
print(type(person_dct))
print(person_dct)
print(person_dct['name'])
```

### Changing dictionary to json: 

```python
import json
# python dictionary
person = {
    "name": "Rushikesh",
    "country": "India",
    "city": "Pune",
    "skills": ["Python", "Machine Learning"]
}


# let's convert it to  json - json.dumps()
person_json = json.dumps(person, indent=1) # indent could be 2, 4, 8. It beautifies the json
print(type(person_json))
print(person_json)
```

### Saving as a json file:

```python
import json

# python dictionary
person = {
    "name": "Rushikesh",
    "country": "India",
    "city": "Pune",
    "skills": ["Python", "Machine Learning"]
}
with open('json_example.json', 'a', encoding='utf-8') as f:
    json.dump(person, f, ensure_ascii=False, indent=4)
```

# Working with xml:

XML is another structured data format which looks like HTML. In XML the tags are not predefined. The first line is an XML declaration. The person tag is the root of the XML. The person has a gender attribute. Example:XML

```python
import xml.etree.ElementTree as ET

tree = ET.parse('example.xml')
parent = tree.getroot()
parent
```

```python
print('Parent tag:', parent.tag)
```

```python
print('Attribute:',  parent.attrib)
```

```python
for child in parent:
    print(child.tag, child.attrib)
```
